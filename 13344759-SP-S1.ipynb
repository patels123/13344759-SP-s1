```python
#decide whether to do outliers treatment or not , if do how?
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
wisker(df['Subscription  Length'])
```

```python
for i in['Subscription  Length','Seconds of Use', 'Frequency of use', 'Frequency of SMS','Customer Value']:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```

```python
#import important Libraries
from sklearn.model_selection import train_test_split 
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
```

```python
df.columns
```

```python
#split data into dependant and independent Features
X = df.drop(["Churn"], axis=1)
y = df[["Churn"]]
```

```python
#split data into testing and training data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 101)
```

```python
#apply standardization
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#create the logistic Regression Model and Test the Model
Model = LogisticRegression()
Model = Model.fit(X_train, y_train)
```

```python
#test the Model
y_pred = Model.predict(X_test)
```

```python
#Check the accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score is", round(accuracy*100),"%")
```

```python
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No Churn', 'Churn'])

# Display the confusion matrix
disp.plot()
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
```

```python
from sklearn.metrics import classification_report

# Print a detailed classification report including precision, recall, and F1-score for each class ('No Churn' and 'Churn')
print(classification_report(y_test, y_pred, target_names=['Churn', 'No Churn']))
```

```python
## Perform hyperparameter tuning using GridSearchCV to find the optimal combination of parameters for a logistic regression model
from sklearn.model_selection import GridSearchCV
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet', None], 
    'C': [0.01, 0.1, 1, 10, 100],                
    'solver': ['lbfgs', 'liblinear', 'saga'],    
    'max_iter': [100, 200, 300]                  
}
```

```python
# Initialize the Logistic Regression model
logistic_model = LogisticRegression()
# Set up GridSearchCV to perform an exhaustive search over the specified parameter grid
grid_search = GridSearchCV(
    estimator=logistic_model, 
    param_grid=param_grid, 
    scoring='accuracy', 
    cv=5,        # 5-fold cross-validation
    verbose=1, 
    n_jobs=-1    # Use all available processors
)
```

```python
# Fit the model on the training data using GridSearchCV to find the best combination of parameters
grid_search.fit(X_train, y_train)

# Print the best hyperparameters found during GridSearchCV
print("Best Parameters:", grid_search.best_params_)

# Print the best cross-validation accuracy score achieved by the best model during GridSearchCV
print("Best Cross-Validation Score:", grid_search.best_score_)
# Test the best model on the test dataset to evaluate its performance on unseen data
best_model = grid_search.best_estimator_  
y_pred = best_model.predict(X_test)       

# Print the accuracy of the model on the test set
print("Test Accuracy:", accuracy_score(y_test, y_pred))

# Print a detailed classification report including precision, recall, and F1-score for each class ('No Churn' and 'Churn')
print(classification_report(y_test, y_pred, target_names=['Churn', 'No Churn']))
```
